SECTION 6: VECTORIZATION (WHY NUMPY IS FAST)

```python
import time

size = 10**6
a = np.random.rand(size)
b = np.random.rand(size)

# Loop method
start = time.time()
result = np.zeros(size)
for i in range(size):
    result[i] = a[i] + b[i]
end = time.time()
print("Loop Time:", end - start)

# Vectorized
start = time.time()
result2 = a + b
end = time.time()
print("Vectorized Time:", end - start)
```

SECTION 7: CONNECTION TO MACHINE LEARNING

```python
# Simple Gradient Descent Example

# Data
x = np.array([1,2,3,4])
y = np.array([2,4,6,8])

# Initialize weight
w = 0.0
learning_rate = 0.01

# Training
for epoch in range(100):
    y_pred = w * x
    loss = np.mean((y - y_pred)**2)

    # derivative
    dw = -2 * np.mean(x * (y - y_pred))

    w = w - learning_rate * dw

print("Final Weight:", w)
```

**WHY DERIVATIVE IS USED IN ML & DL**

Every model has a Loss Function

We need to minimize it

Derivative gives direction of change

Gradient Descent updates parameters

Backpropagation uses chain rule

Without derivatives:

Neural networks cannot learn

Optimization is impossible
